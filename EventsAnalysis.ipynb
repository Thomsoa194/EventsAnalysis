### Number of attendees at each event

```python
plt.figure(figsize=(10, 8))
plt.title('Number of attendees at each event')
plt.xlabel('Number of Attendees')
plt.ylabel('Event')
events_by_attendees = pd.read_csv('Eventsbyattendees.csv')
sns.barplot(x = 'Number of Attendees', y = 'EventName', data = events_by_attendees, palette='pastel');
events_by_attendees['Number of Attendees'].mean()
plt.savefig('How many people attended the event?', bbox_inches='tight')
```

### What is the mean amount per event?

### What was the total generated per event?

```python
from textblob import TextBlob
events_surveys
responses =  events_surveys['Event Experience'].tolist()
responses = ' '.join(responses)
sentiments = TextBlob(responses)
sentiments.sentiment
```

* Using an algorithm for detecting based on a lexicon of predefined words and their score, the polarity score for the 2019 summit is 0.37 and the subjectivity score is 0.58
* Polarity score is measured on a scale of -1 to 1, where -1 is very negative, 0 is neutral and 1 is very positive.
* Subjectivity score is from 0 to 1, where 0 is very objective and 1 is very subjective.
* Therefore, the Events Experience survey responses for the 2019 summit are quite positive and rather subjective.
